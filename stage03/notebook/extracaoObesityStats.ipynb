### Deletar todas as linhas que não são de 2012 até 2014

```
with open('ObesityStats.csv', 'r') as file_in, open('EditYear.csv', 'w') as file_out:
    read = csv.reader(file_in, delimiter=',')
    write = csv.writer(file_out)
    for row in read:
        if (row[0]=='YearStart'):
            write.writerow(row)
        elif ((int(row[0]) >= 2012) and (int(row[0]) <= 2014)):
            write.writerow(row)
```

### Eliminar linhas com dados de idade abaixo de 35 anos e com informação sobre renda e escolaridade.

```
with open('EditYear.csv', 'r') as file_in, open('EditAttr.csv', 'w') as file_out:
    read = csv.reader(file_in, delimiter=',')
    write = csv.writer(file_out)
    for first in read:
        write.writerow(first)
        break
    for row in read:
        shouldWrite = True
        if (row[18] == '18 - 24' or row[18] == '25 - 34'):
            shouldWrite = False 
        if (row[19] != ''):
            shouldWrite = False
        if (row[21] != ''):
            shouldWrite = False
        if (shouldWrite):
            write.writerow(row)
```

### Obter linhas com anos e estados avulsos e juntá-los.

```
states = ['US', 'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA',
          'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 
          'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ', 'NM', 'NY',
          'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
          'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY']

with open('EditAttr.csv', 'r') as file_in, open('togYear.csv', 'w') as file_out:
    read = csv.reader(file_in, delimiter=',')
    write = csv.writer(file_out)
    for first in read:
        write.writerow(first)
        break
    for state in states:
        #2012
        for row in read:
            if (row[2] == state):
                if (row[0] == '2012'):
                    write.writerow(row)
        file_in.seek(0)
        #2013
        for row in read:
            if (row[2] == state):
                if (row[0] == '2013'):
                    write.writerow(row)
        file_in.seek(0)
        #2014
        for row in read:
            if (row[2] == state):
                if (row[0] == '2014'):
                    write.writerow(row)
        file_in.seek(0)
```

### Obter dados só do ano de 2013.

```
with open('togYear.csv', 'r') as file_in, open('ObesityStats_2013.csv', 'w') as file_out:
    read = csv.reader(file_in, delimiter=',')
    write = csv.writer(file_out)
    for row in read:
        if (row[0]=='YearStart'):
            write.writerow(row)
        elif (row[0]=='2013'):
            write.writerow(row)
```
