```
import io
import requests
import json
import csv
```

### Baixar o Dataset e convertê-lo para o formato json do Python.

```
url_json = "https://chronicdata.cdc.gov/views/r35g-znws/rows.json?accessType=DOWNLOAD"
js_json = requests.get(url_json)
json_heart = js_json.json()
```

### Vizualização prévia dos dados.

```
for i in json_heart['data']:
  print(i)
```

```
with open('HeartDisease.csv', 'w') as file_out:
    write = csv.writer(file_out)
    first_row = ['Year', 'LocationAbbr', 'LocationDesc', 'GeographicLevel',
                 'DataSource', 'Class', 'Topic', 'Data_Value', 'Data_Value_Unit',
                 'Data_Value_Type', 'Data_Value_Footnote_Symbol', 'Data_Value_Footnote',
                 'StratificationCategory1', 'Stratification1', 'StratificationCategory2',
                 'Stratification2', 'TopicID', 'LocationID', 'Location 1', 'Counties','States']
    write.writerow(first_row)
    for line in json_heart['data']:
        write.writerow(line[8:])
```
